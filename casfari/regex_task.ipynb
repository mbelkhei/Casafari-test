```python
import codecs
from bs4 import BeautifulSoup as bs
import re
import json
import textwrap
import os
```

```python
path_html_file = str(input('type the path to your html file: ')) # path of file. Defaut is within the the same file as the script
html_file = str(input('type the name of your html file: ')) # suggestion: sample.html

path_output_file = str(input('type the path to your output file: ')) # path of output file. Defaut is within the the same file as the script
output_file = str(input('type the name of your text file: ')) # suggestion: 'regex_output.txt' 

def check_path(path):
    if path[-1] == '\\':
        return path
    else:
        path = path+'\\'
        #print(path)
        return path



path_html_file = check_path(path_html_file)
path_output_file = check_path(path_output_file)

try:
    doc = codecs.open(path_html_file+html_file) # load the file
except: 
    print('file not found in path')

sp = bs(doc.read(), 'html.parser') # parse the file
text = sp.prettify() # html as string

output_text= open(path_output_file+output_file,'w')
```

```python
# extract links using beautifulsoup
output_text.write('1)Extraction of links using beautifulsoup\n'+50*'-'+'\n')
for i in sp.find_all('a'):
    output_text.write(i.get('href')+'\n')
    print(i.get('href'))
output_text.write(50*'-'+'\n\n')
```

```python
# extract links using regex
output_text.write('2)Extraction of links using regex\n'+50*'-'+'\n')
links = re.findall("<a[^>]+href=\"(.*?)\"", text)
for link in links:
    output_text.write(link+'\n')
    print(link)
output_text.write(50*'-'+'\n\n')
```

```python
# Extraction of latitude and longitude using regex
output_text.write('3)Extraction of latitude and longitude using regex\n'+50*'-'+'\n')
latlong = re.findall('\"(\-?\d+\.\d+)\,(\-?\d+\.\d+)\"',text)[0]
latitude = latlong[0]
longitude = latlong[1]
output_text.write('Latitude is: {}\nLongitude is: {}\n'.format(latitude,longitude))
output_text.write(50*'-'+'\n\n')
```

```python
# read json file
path_json_file = str(input('type the path to your json file: ')) # path of file. Defaut is within the the same file as the script
json_file = str(input('type the name of your json file: ')) # suggestion: sample.json

path_json_file = check_path(path_json_file)
with open(json_file) as f:
    data = json.load(f)
    f.close()
json_string = json.dumps(data)
```

```python
uniqueID_askey = list(set(data['objekte']['objekte'].keys())) # using json dict to extract unique IDs
print(uniqueID_askey)
```

```python
output_text.write('4)Extraction of unique IDs using regex\n'+50*'-'+'\n')
list_of_unique_IDs_from_string = list(set(re.findall('[A-Z]+[0-9]+', json_string))) # using json string with regex to extract unique ID
t_id = ','.join(list_of_unique_IDs_from_string) 
output_text.write('['+textwrap.fill(t_id, 48)+']\n')
output_text.write(50*'-'+'\n\n')
print(list_of_unique_IDs_from_string)
```

```python
output_text.write('5)Extraction of unique IDs that contain ESP lands\nusing regex\n'+50*'-'+'\n')
list_of_unique_IDs_from_string_with_ESPland= list(set(re.findall('\"([A-Z]+[0-9]+)\".*?\"land\":\s\"ESP\".*?', json_string)))
t_id_land = ','.join(list_of_unique_IDs_from_string_with_ESPland) 
output_text.write('['+textwrap.fill(t_id_land, 48)+']\n')
output_text.write(50*'-'+'\n\n')
print(list_of_unique_IDs_from_string_with_ESPland)
```
